#### Create a correlation heatmap using seaborn:

This correlation heatmap uses color and coefficients to indicate the strength of a correlation. The coefficient ranges are from -1 to 1 with -1 being a strong negative relationship and anywhere from 0.5-1 being a strong positive relationship. The legend indicates purple equates to 0, or no relationship between the variables, dark purple and black equate to negative correlations, and orange and light orange equate to positive correlations.

n_injured and victim_count have a strong positive relationship with a coefficient of 0.69. This indicates that as one variable increases, the other does as well. 

Other pairings that show strong positive relationships include: n_gun_involved and unknown_gun_count with a coefficient of 0.96, suspect_count and male_count at 0.66, and adult_count and male_count at 0.74.

### Categorical Plots:

```python
# Create a new DataFrame with the total deaths for each gun type
df_long = pd.DataFrame({
    'pistol_deaths': df['n_killed'] * df['pistol_count'],
    'shotgun_deaths': df['n_killed'] * df['shotgun_count'],
    'rifle_deaths': df['n_killed'] * df['rifle_count'],
    'unknown_deaths': df['n_killed'] * df['unknown_gun_count']
})

# Convert to long format for Seaborn plotting
df_long_melt = df_long.melt(var_name='weapon_type', value_name='n_killed')

# Rename the weapon_type values to be more readable
df_long_melt['weapon_type'] = df_long_melt['weapon_type'].replace({
    'pistol_deaths': 'Pistol',
    'shotgun_deaths': 'Shotgun',
    'rifle_deaths': 'Rifle',
    'unknown_deaths': 'Unknown'
})

# Group by weapon_type and sum the number of deaths
total_n_killed = df_long_melt.groupby('weapon_type')['n_killed'].sum().reset_index()

# Custom colors for each weapon type
custom_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # blue, orange, green, red for unknown

# Plot using catplot and apply custom colors directly
g = sns.catplot(x='weapon_type', y='n_killed', data=total_n_killed, kind='bar', height=5, aspect=1.5)

# Add count annotations to the bars
for ax in g.axes.flatten():
    for p, color in zip(ax.patches, custom_colors):
        p.set_facecolor(color)  # Set the color for each bar
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='bottom', fontsize=10)

plt.title('Total Number of People Killed in the U.S. by Gun Type (2013-2018)')
plt.ylabel('Number of Deaths')
plt.xlabel('Gun Type')
plt.xticks(rotation=45)
plt.show()
```

This categorical bar plot shows the total number of people killed in the U.S. by gun type (pistol, rifle, shotgun, or unknown gun type) from the years 2013-2018. I am surprised and a bit disappointed that the "Unknown" gun type is the highest. This could be due to several reasons such as lack of reporting, unregistered firearms, and missing data. The rest of the data looks accurate upon crosschecking.
